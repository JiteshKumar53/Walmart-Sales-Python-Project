```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:\\Users\\Jitesh Kumar\\Downloads\\Walmart_Project\\Walmart.xlsx')
df.info()
```

```python
df
```

```python
#convert store into string

df['Store']=df['Store'].astype(str)
```

```python
average_weekly_sales_by_store=df.groupby('Store')['Weekly_Sales'].mean().reset_index()
average_weekly_sales_by_store['Weekly_Sales'] /= 1000000
average_weekly_sales_by_store
```

```python
plt.figure(figsize=(20,6))
import matplotlib.pyplot as plt
plt.bar(average_weekly_sales_by_store['Store'],average_weekly_sales_by_store['Weekly_Sales'],color='green')
plt.title('Weekly Sales Chart')
plt.xlabel('Stores')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
df_sort = average_weekly_sales_by_store.sort_values(by='Weekly_Sales',ascending=False)
df_sort
```

```python
Top_5_stores=df_sort.head(5)
Top_5_stores
```

```python
Bottom_5_stores=df_sort.head(5)
Bottom_5_stores
```

```python
plt.figure(figsize=(8,3))
plt.bar(Top_5_stores['Store'],Top_5_stores['Weekly_Sales'],color='g')
plt.title('Top 5 Stores')
plt.xlabel('Stores')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.bar(Bottom_5_stores['Store'],Bottom_5_stores['Weekly_Sales'],color='r')
plt.title('Bottom 5 stores')
plt.xlabel('Sales')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
#Bar chat as per Holiday Sales
average_weekly_sales_by_Holiday= df.groupby('Holiday_Flag')['Weekly_Sales'].mean().reset_index()
average_weekly_sales_by_Holiday['Weekly_Sales']/=1000000
average_weekly_sales_by_Holiday
```

```python
plt.figure(figsize=(6,4))
plt.bar(average_weekly_sales_by_Holiday['Holiday_Flag'],average_weekly_sales_by_Holiday['Weekly_Sales'])
plt.title('Sales on Holidays')
plt.xlabel('')
plt.xticks([0,1],['No Holiday','Holiday'])
plt.ylabel('Weekly Sales')
plt.show()
```
